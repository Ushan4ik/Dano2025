```python
def remove_outliers_iqr(df, columns, whis):
    df_clear = df.copy()
    for col in columns:
        Q1 = df_clear[col].quantile(0.25)
        Q3 = df_clear[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - whis * IQR
        upper = Q3 + whis * IQR
        df_clear = df_clear[
            (df_clear[col] >= lower) &
            (df_clear[col] <= upper)
        ]
    return df_clear
```

```python
def calc_hhi(df):
    hhi_bud = {}
    hhi_area = {}
    for i in df["area"].unique():
        hhi_bud[i] = 0
        hhi_area[i] = 0
    for i in df["area"].unique():
        df1 = df[df["area"] == i]
        all1 = 0
        all2 = 0
        summ1 = {}
        summ2 = {}
        for j in df1["developer"].unique():
            summ1[j] = 0
            summ2[j] = 0
        for j in df1["developer"].unique():
            summ1[j] = df1[df1["developer"] == j]["total_budget"].sum()
            summ2[j] = df1[df1["developer"] == j]["total_deal_area"].sum()
            all1 += summ1[j]
            all2 += summ2[j]
        for j in summ1.keys():
            summ1[j] /= all1
            summ2[j] /= all2
        for j in summ1.keys():
            hhi_bud[i] += (summ1[j] * 100) ** 2
            hhi_area[i] += (summ2[j] * 100) ** 2
    #print("--- Budget HHI ---")
    #for i in hhi_bud.keys():
        #print(f"{i}:", round(hhi_bud[i]))
    #print("\n--- Deal area HHI ---")
    #for i in hhi_area.keys():
        #print(f"{i}:", round(hhi_area[i]))
    return hhi_bud, hhi_area
hhi_bud, hhi_area = calc_hhi(df_trades_clr)
```

```python
build_vel = {}
build_vel_of_all = {}
for i in df_trades_clr["building_id"].unique():
    build_vel[i] = 0
    build_vel_of_all[i] = 0
for i in df_trades_clr["building_id"].unique():
    df1 = df_trades_clr[df_trades_clr["building_id"] == i]
    min_date = df1["contract_date"].min()
    max_date = df1["contract_date"].max()
    active = max((max_date - min_date).days / 30, 1)
    summ = df1["deals_count"].sum()
    build_vel_of_all[i] = summ / all[i] / active
    build_vel[i] = summ / active
```

Интерпретация результатов теста Шапиро-Уилка в статистике заключается в оценке соответствия распределения данных нормальному распределению. Тест проверяет нулевую гипотезу: данные распределены нормально, и альтернативную — данные не имеют нормального распределения.



```python
from scipy import stats
print("Low:", "Yes" if stats.shapiro(values_comp["Low"]).pvalue >= 0.05 else "No")
print("Mid:", "Yes" if stats.shapiro(values_comp["Mid"]).pvalue >= 0.05 else "No")
print("High:", "Yes" if stats.shapiro(values_comp["High"]).pvalue >= 0.05 else "No")
```

Тест Левене (Levene's Test) в статистике используется для оценки равенства дисперсий для переменной, рассчитанной для двух или более групп. Тест проверяет нулевую гипотезу о том, что дисперсии совокупности равны (однородность дисперсии или гомоскедастичность).

```python
print("Levene:", "Yes" if stats.levene(values_comp["Low"], values_comp["Mid"], values_comp["High"]).pvalue >= 0.05 else "No")
```

Используем тест Краскелаа-Уолиса

```python
kruskal = stats.kruskal(values_comp["Low"], values_comp["Mid"], values_comp["High"])
print("--- Kruskal:\n", kruskal, sep='')
```

```python
print("Effect size of Kruskal:", kruskal.statistic / len(values_comp["Low"] + values_comp["Mid"] + values_comp["High"]))
```

P-value < 0.05. Медианы различаются!

Также размер эффекта ~= 0.285. Это много! Это означает, что уровень конкуренции между застройщиками объясняет приблизительно 28.5% вариации цен на первичном рынке новостроек Новой Москвы.

```python
for room in ['ст', 1, 2, 3]:
    df_work = df_trades_clr[df_trades_clr["rooms"] == room].copy()
    print("Testing", room, "room only...")
    hhi_bud, hhi_area = calc_hhi(df_work)
    plt.bar(hhi_bud.keys(), hhi_bud.values(), label='Budget', zorder=0)
    plt.bar(hhi_area.keys(), hhi_area.values(), label='Deal area', zorder=1, color='yellow', alpha=0.3)
    plt.axhline(y=np.median(list(hhi_area.values())))
    plt.legend(loc='upper left')
    plt.title(f"Budget and deal area HHI per region({room} room only!)");
    plt.show()
    df_hhi = pd.DataFrame(index=hhi_bud.keys(), columns=["HHI"], data=hhi_bud.values())
    if(room == 3):
        df_hhi["Seg"] = pd.qcut(df_hhi["HHI"], [0, 0.25, 0.75, 1], labels=["Low", "Mid", "High"])
    else:
        df_hhi["Seg"] = pd.qcut(df_hhi["HHI"], 3, labels=["Low", "Mid", "High"])
    print(df_hhi["Seg"])
    comp = {"Low" : [], "Mid" : [], "High" : []}
    for i in df_hhi.index:
        comp[df_hhi.loc[i, "Seg"]].append(i)
    values_comp = {"Low" : [], "Mid" : [], "High" : []}
    for i in comp.keys():
        df1 = df_work[df_work["area"].isin(comp[i])]
        values3 = []
        for j in df1["building_id"].unique():
            summ = 0
            df2 = df1[df1["building_id"] == j]
            for k in df2.index:
                summ += (df2.loc[k, "total_budget"] / df2.loc[k, "total_deal_area"])
            values3.append(summ / len(df2))
        values_comp[i] = values3
    print("Low:", len(values_comp["Low"]))
    print("Mid:", len(values_comp["Mid"]))
    print("High:", len(values_comp["High"]))
    print("--- Shapiro ---")
    print("Low:", "Yes" if stats.shapiro(values_comp["Low"]).pvalue >= 0.05 else "No")
    print("Mid:", "Yes" if stats.shapiro(values_comp["Mid"]).pvalue >= 0.05 else "No")
    print("High:", "Yes" if stats.shapiro(values_comp["High"]).pvalue >= 0.05 else "No")
    print("--- Levene ---")
    print("Yes" if stats.levene(values_comp["Low"], values_comp["Mid"], values_comp["High"]).pvalue >= 0.05 else "No")
    kruskal = stats.kruskal(values_comp["Low"], values_comp["Mid"], values_comp["High"])
    print("--- Kruskal ---\n", kruskal, sep='')
    print("Effect size of Kruskal:", kruskal.statistic / len(values_comp["Low"] + values_comp["Mid"] + values_comp["High"]))
    print("--- Dann ---\n")
    df_dunn = pd.DataFrame(columns=["comp_segment", "metre_price"])
    df_dunn["comp_segment"] = ["Low"] * len(values_comp["Low"]) + ["Mid"] * len(values_comp["Mid"]) + ["High"] * len(values_comp["High"])
    df_dunn["metre_price"] = list(values_comp["Low"]) + list(values_comp["Mid"]) + list(values_comp["High"])
    dunn_test = sp.posthoc_dunn(df_dunn, val_col='metre_price', group_col='comp_segment', p_adjust='holm')
    for i in dunn_test.columns:
        dunn_test[i] = dunn_test[i].apply(lambda x: "YES" if x < 0.05 else "NO")
    print(dunn_test)
```
